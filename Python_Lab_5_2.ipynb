```python
import random

def tao_danh_sach_ngau_nhien(n):
    danh_sach = []
    for i in range(n):
        danh_sach.append(random.randint(1, 100))
    return danh_sach

def in_danh_sach(danh_sach):
    print("Danh sách:", danh_sach)

def in_dao_nguoc(danh_sach):
    print("Danh sách đảo ngược:", danh_sach[::-1])

def sap_xep_va_in(danh_sach):
    danh_sach_da_xep = sorted(danh_sach)
    print("Danh sách đã sắp xếp:", danh_sach_da_xep)

def tim_phan_tu_lon_nhat(danh_sach):
    lon_nhat = max(danh_sach)
    vi_tri = len(danh_sach) - 1 - danh_sach[::-1].index(lon_nhat)
    print(f"Phần tử lớn nhất: {lon_nhat}, vị trí cuối cùng: {vi_tri}")

def dem_phan_tu_co_gia_tri_x(danh_sach, x):
    vi_tri = [i for i, phan_tu in enumerate(danh_sach) if phan_tu == x]
    print(f"Số lượng phần tử có giá trị {x}: {danh_sach.count(x)}")
    print(f"Vị trí xuất hiện: {vi_tri}")

def in_so_nguyen_to(danh_sach):
    def la_so_nguyen_to(n):
        if n < 2:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

    vi_tri = [i for i, phan_tu in enumerate(danh_sach) if la_so_nguyen_to(phan_tu)]
    print(f"Vị trí các số nguyên tố: {vi_tri}")

def tim_so_duy_nhat(danh_sach):
    so_duy_nhat = [x for x in danh_sach if danh_sach.count(x) == 1]
    print(f"Các số xuất hiện duy nhất: {so_duy_nhat}")

def liet_ke_tan_suat(danh_sach):
    tan_suat = {}
    for phan_tu in danh_sach:
        if phan_tu in tan_suat:
            tan_suat[phan_tu] += 1
        else:
            tan_suat[phan_tu] = 1

    print("Giá trị và số lần xuất hiện:")
    for phan_tu, so_lan in tan_suat.items():
        print(f"{phan_tu}: {so_lan} lần")

def tim_doan_giam_lien_tiep(danh_sach):
    if len(danh_sach) < 2:
        print("Danh sách quá ngắn để tìm đoạn giảm")
        return

    print("Các đoạn con giảm liên tiếp:")
    bat_dau = 0
    for i in range(1, len(danh_sach)):
        if danh_sach[i] >= danh_sach[i-1]:
            if i - bat_dau > 1:
                print(danh_sach[bat_dau:i])
            bat_dau = i

    if len(danh_sach) - bat_dau > 1:
        print(danh_sach[bat_dau:])

def main():
    while True:
        try:
            n = int(input("Nhập số lượng phần tử N: "))
            if n <= 0:
                print("Vui lòng nhập số nguyên dương.")
                continue
            break
        except ValueError:
            print("Vui lòng nhập một số nguyên dương.")

    danh_sach = tao_danh_sach_ngau_nhien(n)

    while True:
        print("1. In ra danh sách")
        print("2. In đảo ngược danh sách")
        print("3. Sắp xếp và in danh sách")
        print("4. Tìm phần tử lớn nhất và vị trí cuối cùng")
        print("5. Đếm phần tử có giá trị X")
        print("6. In vị trí các số nguyên tố")
        print("7. Tìm các số duy nhất")
        print("8. Liệt kê tần suất xuất hiện")
        print("9. In các đoạn con giảm liên tiếp")
        print("0. Thoát")

        lua_chon = input("Nhập lựa chọn của bạn: ")
        print()

        if lua_chon == '1':
            in_danh_sach(danh_sach)
        elif lua_chon == '2':
            in_dao_nguoc(danh_sach)
        elif lua_chon == '3':
            sap_xep_va_in(danh_sach)
        elif lua_chon == '4':
            tim_phan_tu_lon_nhat(danh_sach)
        elif lua_chon == '5':
            try:
                x = int(input("Nhập giá trị X cần đếm: "))
                dem_phan_tu_co_gia_tri_x(danh_sach, x)
            except ValueError:
                print("Vui lòng nhập một số nguyên.")
        elif lua_chon == '6':
            in_so_nguyen_to(danh_sach)
        elif lua_chon == '7':
            tim_so_duy_nhat(danh_sach)
        elif lua_chon == '8':
            liet_ke_tan_suat(danh_sach)
        elif lua_chon == '9':
            tim_doan_giam_lien_tiep(danh_sach)
        elif lua_chon == '0':
            print("Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")

if __name__ == "__main__":
    main()
```

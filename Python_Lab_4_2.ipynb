```python
import random
import time

def game_tai_xiu():
    tien = 100000
    so_van_thang = 0
    so_van_choi = 0

    while tien > 0:
        print(f"\nSố tiền hiện tại của bạn: {tien:,} VND")

        while True:
            try:
                tien_cuoc = int(input(f"Nhập số tiền cược (tối đa {tien:,} VND): "))
                if tien_cuoc <= 0:
                    print("Số tiền cược phải lớn hơn 0!")
                elif tien_cuoc > tien:
                    print(f"Bạn không đủ tiền! Bạn chỉ có {tien:,} VND.")
                else:
                    break
            except ValueError:
                print("Vui lòng nhập một số hợp lệ!")

        while True:
            lua_chon = input("\nBạn muốn đặt cược vào (Tài/Xỉu/5): ").lower()
            if lua_chon in ["tài", "tai", "t", "xỉu", "xiu", "x", "5"]:
                break
            else:
                print("Lựa chọn không hợp lệ! Vui lòng chọn 'Tài', 'Xỉu' hoặc '5'.")

        if lua_chon in ["tài", "tai", "t"]:
            lua_chon_chuan = "Tài"
        elif lua_chon in ["xỉu", "xiu", "x"]:
            lua_chon_chuan = "Xỉu"
        else:
            lua_chon_chuan = "5"

        print(f"\nBạn đã đặt cược {tien_cuoc:,} VND vào {lua_chon_chuan}")
        print("\nGieo xúc sắc")

        xuc_sac_1 = random.randint(1, 6)
        xuc_sac_2 = random.randint(1, 6)
        tong_xuc_sac = xuc_sac_1 + xuc_sac_2

        print(f"Xúc sắc 1: {xuc_sac_1}")
        print(f"Xúc sắc 2: {xuc_sac_2}")
        print(f"Tổng: {tong_xuc_sac}")

        # Xác định kết quả
        if tong_xuc_sac > 5:
            ket_qua = "Tài"
        else:
            ket_qua = "Xỉu"

        # Kiểm tra người chơi thắng hay thua
        if lua_chon_chuan == "5" and tong_xuc_sac == 5:
            tien += tien_cuoc * 3
            so_van_thang += 1
            print(f"\nChúc mừng! Bạn đã thắng với số 5 chính xác!")
            print(f"Bạn nhận được {tien_cuoc * 3:,} VND!")
        elif lua_chon_chuan == ket_qua:
            tien += tien_cuoc
            so_van_thang += 1
            print(f"\nChúc mừng! Kết quả là {ket_qua}!")
            print(f"Bạn thắng {tien_cuoc:,} VND!")
        else:
            tien -= tien_cuoc
            print(f"\nRất tiếc! Kết quả là {ket_qua}!")
            print(f"Bạn thua {tien_cuoc:,} VND!")

        so_van_choi += 1

        if tien <= 0:
            print("\nBạn đã hết tiền! Game kết thúc.")
            break

        lua_chon = input("\nBạn có muốn chơi tiếp không? (có/không): ").lower()
        if lua_chon not in ["có", "co", "c", "yes", "y"]:
            break

    print("\nThống kê:")
    print(f"Số tiền còn lại: {tien:,} VND")
    print(f"Số ván thắng: {so_van_thang}")
    print(f"Tổng số ván đã chơi: {so_van_choi}")

    if tien > 100000:
        print(f"Lợi nhuận: +{tien - 100000:,} VND")
    elif tien < 100000:
        print(f"Lỗ vốn: -{100000 - tien:,} VND")
    else:
        print("Hòa vốn: 0 VND")

    print("\nCảm ơn bạn đã chơi game Tài Xỉu! Tạm biệt.")

# Chạy game
if __name__ == "__main__":
    game_tai_xiu()
```
